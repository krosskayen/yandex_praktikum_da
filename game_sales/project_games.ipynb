# Определение успешности продаж компьютерных игр

Вам предоставлен файл из открытых источников с историческими данными о продажах компьютерных игр за период до 2016 года. Необходимо проанализировать предоставленные данные и выявить закономерности, определяющие успешность игры. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламную компанию на 2017 год. 

## Изучение общей информации по предоставленным данным

```python
# импортирование необходимых библиотек
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
sns.set_style("whitegrid")
%config InlineBackend.figure_format='retina'
%matplotlib inline

from scipy import stats as st
from tqdm import tqdm

import warnings
warnings.filterwarnings(action='ignore')

# укажем автоматическое округление формата float до 3 цифр после запятой
pd.set_option('display.float_format', lambda x: '%.3f' % x)

# укажем стандартные размеры графиков в дюймах
rcParams['figure.figsize'] = 15, 7.5

# снимем ограничение на количество отображаемых столбцов
pd.set_option('display.max_columns', None)
```

Прочитаем исходные файлы с данными и сохраним их в переменную, затем ознакомимся.

```python
df = pd.read_csv('games.csv')
```

```python
display(df.info())
display(df.head())
```

Предоставленная таблица `games` состоит из 16715 строк и 11 столбцов с данными о продаже игр. К каждому из столбцов имеется описание содержащихся в них данных: 
* `Name` — название игры;
* `Platform` — платформа;
* `Year_of_Release` — год выпуска;
* `Genre` — жанр игры;
* `NA_sales` — продажи в Северной Америке (миллионы проданных копий);  
* `EU_sales` — продажи в Европе (миллионы проданных копий);
* `JP_sales` — продажи в Японии (миллионы проданных копий);
* `Other_sales` — продажи в других странах (миллионы проданных копий);
* `Critic_Score` — оценка критиков (максимум 100);
* `User_Score` — оценка пользователей (максимум 10);
* `Rating` — возрастной рейтинг от организации ESRB (англ. Entertainment Software Rating Board).

Столбцы имеют в целом корректное наименование. В процессе подготовки данных стоит привести их к нижнему регистру, а типы данных привести к корректным.
В нескольких столбцах имеются пропущенные значения, однако исходя из предоставленного описания к предоставленной таблице, данные за 2016 год могут быть неполными, необходимо подробнее с ними ознакомиться.

### Предобработка данных

Для начала изменим наименование столбцов, в верхнем регистре на нижний.

Предварительно мы установили, что в нескольких столбцах имеется большое количество пропусков, посмотрим какую часть они занимают относительно всей выборки, а также посмотрим что можно сделать для их устранения.
После чего приведем данные в столбцах к корректному типу для проведения дальнейших преобразований и дальнейшего исследования.

```python
df.columns = df.columns.str.lower()
```

```python
df.head()
```

Заполнить данные в столбцах какими-либо значениями не представляется возможным. Поскольку данные датированы 1993 годом и в них не указан какой-либо жанр, они бесполезны для нашего исследования, их можно удалить.

Посмотрим на значения каждого столбца подробнее.

```python
df.year_of_release.value_counts()
```

```python
df.genre.value_counts()
```

Значения в столбце с наименованием жанра вполне корректны, какие-либо исправления не требуются. 

Пропущенные значения в столбце с годом релиза вероятнее всего появились вследствие некорректного заполнения данных в базе (так например, некоторые названия уже содержат год релиза). Теоретически можно было бы заполнить данные в столбце корректными значениями, но это займет неоправданно большое количество времени. 

Строки с данными, в которых имеются пропущенные значения в столбце с годом релиза составляют небольшое количество (1.6%), а поскольку нам ставится задача прогнозирования продаж на конкретный год (2017), в связи с этим указанные данные бесполезны для нашего исследования, удалим их.

```python
df = df.dropna(subset=['year_of_release'], axis=0).reset_index(drop=True)
print(df[['year_of_release']].isna().sum())
```

```python
df.critic_score.value_counts()
```

Данные вполне корректны, исходя из имеющихся значений в столбце, тип данных в столбце с оценками критиков можно заменить на int.

```python
df.user_score.value_counts()
```

Большое количество игр (2424) имеет рейтинг пользователей 'tbd', что расшифровывается как 'To Be Determined'. Буквально это означает, 'будет определено', т.е. пользователи еще не ставили оценки этой игре. Для приведения типа данных в столбце к корректному типу, необходимо будет избавиться от этого значения, заменив его на NaN. Это не помешает нам в исследовании, т.к. соответствующий тип корректно обрабатывается при проведении с ним математических операций. После чего, изменим тип данных в столбце с object на float.

```python
df.loc[(df['user_score'] == 'tbd'), 'user_score'] = np.nan
```

```python
df.user_score.value_counts()
```

```python
df.rating.value_counts()
```

Большое количество незаполненных значений в столбцах с оценкой пользователей и оценкой критиков предположительно объясняется тем, что далеко не всем играм (особенно имеющим незначительное число продаж), пользователи/критики оставляют какую-либо оценку, либо эта оценка берется из каких-то единственных источников (а не большого количества тематических сайтов/форумов/журналов с рейтингами по всем странам).

Что касается пропусков в столбце с данными о рейтинге, то согласно данным из открытых источников Entertainment Software Rating Board (ESRB) - негосударственная организация, основное направление деятельности которой состоит в принятии и определении рейтингов для компьютерных видеоигр и другого развлекательного программного обеспечения в США и Канаде, т.е. если игра не поставлялась на рынки этих двух стран, рейтинг ей мог и не присваиваться.

## Исследовательский анализ данных

### Распределение количества выпускаемых игр по годам выпуска

Создадим таблицу с данными по году выпуска и количеству игр, выпущенных в указаный период, затем построим по ней диаграмму. 

```python
games_for_years = df.pivot_table(index='year_of_release', values='name', aggfunc='count').reset_index()
games_for_years.columns = ['year', 'count_games']
display(games_for_years.head())
```

```python
(sns.barplot(x='year', y='count_games', palette='deep', 
    data=games_for_years) 
    .set(title='Общее количество выпущенных игр по годам', 
    xlabel='Год выпуска', 
    ylabel='Количество игр (в штуках)'))
plt.xticks(rotation='vertical')
plt.show()

```

Количество выпускаемых игр планомерно увеличивалось примерно с начала-середины 90-х годов ХХ века и достигло своего пика в 2008-2009 годах. После чего, примерно с 2012 года заметен довольно резкий спад в количестве выпускаемых игр, с 2012 по 2016 годы продажи игр держатся примерно на одном уровне.

Для дальнейшего исследования пригодятся данные по продажам за период с 2012 по 2016 год. Данные за более ранние периоды могут значительно исказить наше исследование, поскольку с 2007 по 2011 годы выпускали значительно больше игр, чем в последующие годы, а одних лишь данных за 2015-2016 годы может быть недостаточно для исследования.

### Распределение количества проданных копий игр по актуальных платформ

Создадим таблицу для платформ с наибольшими суммарными продажами по каждой из платформ, а затем построим диаграмму и посмотрим на распределение продаж для платформ.

```python
platforms = (df.pivot_table(index='platform', values='total_sales', aggfunc='sum')
            .sort_values(by='total_sales', ascending=False).reset_index())
display(platforms.head(10))
```

```python
(sns.barplot(x='platform', y='total_sales', palette='deep', 
    data=platforms.sort_values(by='total_sales')) 
    .set(ylim=(0,1400), title='Общее количество проданных копий игр для каждой платформы', 
    xlabel='Платформа', 
    ylabel='Количество проданнных игр (в млн. штук)'))
plt.xticks(rotation='vertical')
plt.show()
```

Из всех платформ по общему количеству проданных копий игр заметно выделяются PS2, X360, PS3, Wii, DS, PS. Построим распределение продаж по годам для каждой из этих платформ.

```python
# выделим эти платформы в отдельный список
best_platforms_all_years = list(platforms.loc[:5, 'platform'])
display(best_platforms_all_years)
```

```python
platforms_per_year = (df[df.platform.isin(best_platforms_all_years)].pivot_table(index=['platform', 'year_of_release'], 
                    values='total_sales', aggfunc='sum').reset_index())
display(platforms_per_year.head())
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5), sharey=True)
fig.suptitle('Количество проданных игр на платформах по годам')

sns.barplot(ax=axes[0], x='year_of_release', y='total_sales', 
palette='deep', data=platforms_per_year.query('platform == "PS2"'))
axes[0].set(title='PS2', xlabel='Год выпуска', ylabel='Количество продаж (в млн. штук)')

sns.barplot(ax=axes[1], x='year_of_release', y='total_sales', 
palette='deep', data=platforms_per_year.query('platform == "X360"'))
axes[1].set(title='X360', xlabel='Год выпуска', ylabel='Количество продаж (в млн. штук)')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5), sharey=True)
fig.suptitle('Количество проданных игр на платформах по годам')

sns.barplot(ax=axes[0], x='year_of_release', y='total_sales', 
palette='deep', data=platforms_per_year.query('platform == "PS3"'))
axes[0].set(title='PS3', xlabel='Год выпуска', ylabel='Количество проданных копий (в млн. штук)')

sns.barplot(ax=axes[1], x='year_of_release', y='total_sales', 
palette='deep', data=platforms_per_year.query('platform == "Wii"'))
axes[1].set(title='Wii', xlabel='Год выпуска', ylabel='Количество проданных копий (в млн. штук)')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5), sharey=True)
fig.suptitle('Количество проданных игр на платформах по годам')

sns.barplot(ax=axes[0], x='year_of_release', y='total_sales', 
palette='deep', data=platforms_per_year.query('platform == "DS"'))
axes[0].set(title='DS', xlabel='Год выпуска', ylabel='Количество продаж (в млн. штук)')

sns.barplot(ax=axes[1], x='year_of_release', y='total_sales', 
palette='deep', data=platforms_per_year.query('platform == "PS"'))
axes[1].set(title='PS', xlabel='Год выпуска', ylabel='Количество продаж (в млн. штук)')
```

Продажи игр в момент появления новой платформы, как правило относительно небольшие. На второй год продажи заметно увеличиваются и, как правило, в промежутке с 2 до 4-5 лет продажи держатся на стабильно высоком уровне, потом идут на спад. Платформы в целом дают продажи в промежутке от 7 до 9 лет, затем количество продаваемых игр на них становится крайне низким.
```python
supposition_1 = supposition_1.dropna(subset=['user_score'], axis=0).reset_index(drop=True)
print(supposition_1.user_score.isna().sum())
```

**Введем основную и альтернативные гипотезы:**

H0: Средние пользовательские рейтинги (оценки) для платформ `Xbox One`и `PC` не различаются.    
H1: Средние пользовательские рейтинги (оценки) для платформ `Xbox One`и `PC` различаются.

Выделим исследуемые признаки в отдельные таблицы.

```python
xbox = supposition_1.query('platform == "XOne"')
pc = supposition_1.query('platform == "PC"')
```

```python
display(xbox.head(1))
display(pc.head(1))
```

Проверим получившиеся таблицы с данными на нормальность распределения при помощи гистограмм и QQ-графиков.

```python
# строим гистограмму
fig, ax = plt.subplots(1, 2)

sns.histplot(x='user_score', data=xbox, kde=True, ax=ax[0])
sns.histplot(x='user_score', data=pc, kde=True, ax=ax[1])
```

```python
# строим qq-plot для нормального распределения
plt.subplot(2,2,1)
st.probplot(xbox['user_score'], dist="norm", plot=plt)
plt.subplot(2,2,2)
st.probplot(pc['user_score'], dist="norm", plot=plt)
plt.show()
```

Распределение значений в обеих выборках с некоторым допущением можно отнести к нормальному, предоставленные выборки независимы. Для проверки равенства средних полагаю возможным использовать t-тест. В то же время, поскольку сами выборки довольно небольшого размера, можно дополнительно вычислить и посмотреть средние значения по каждой выборке.

Посчитаем p-value, используя ttest из библиотеки scipy для сравнения средних двух выборок.

Зададим уровень значимости: 𝛼 = 0.05 

Если p-value будет меньше выбранного уровня значимости 𝛼 = 0.05, то нулевую гипотезу об отсутствии отличий средних отвергнем. Иначе сделаем вывод, что данные не позволили нам отвергнуть нулевую гипотезу.

```python
# зададим критический уровень статистической значимости
alpha_1 = .05

results_1 = st.ttest_ind(xbox.user_score, pc.user_score, equal_var=False)

print('p-значение: ', results_1.pvalue)

if (results_1.pvalue < alpha_1):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
    
print('Средние оценки пользователей для платформы "Xbox One":', xbox.user_score.mean())
print('Средние оценки пользователей для платформы "PC":', pc.user_score.mean())
```

Таким образом, на имеющихся данных, на уровне значимости 5% нет оснований отвергнуть нулевую гипотезу, согласно которой cредние пользовательские оценки для платформ `Xbox One`и `PC` не различаются.
