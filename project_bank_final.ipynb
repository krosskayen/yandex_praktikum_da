### Изучение распределения признаков в разрезе оттока

```python
# столбцы с признаками, имеющими биноминальное значение или категориальными
binominal = ['city', 'gender', 'equity', 'products', 'credit_card', 'last_activity']
```

```python
def graf_for_binominal_churn(data, column):   
    pallete = sns.color_palette("tab10", 7)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (14,6), gridspec_kw={'width_ratios': [1.5, 1.5]})
    
    sns.histplot(data=df.query('churn == False')[column], color=pallete[0], 
                 stat='probability', common_norm=False, ax=ax1, bins=df[column].nunique()) 
    sns.histplot(data=df.query('churn == True')[column], color=pallete[1], 
                 stat='probability', common_norm=False, ax=ax1, bins=df[column].nunique())
    ax1.set_ylabel('Частота признака')
    ax1.set_xlabel(''.format(column))
    ax1.set_title(f'Распределение признака {column}', fontsize=12, fontweight='bold')
    ax1.legend(labels=['Остались', 'Ушли'])
    
    sns.countplot(df[column], hue=df['churn'], ax=ax2, palette="deep")
    ax2.set_ylabel('Количество клиентов')
    ax2.set_title(f'Распределение признака {column}', fontsize=12, fontweight='bold')
    ax2.set_xlabel('')
    ax2.legend(labels=['Остались', 'Ушли'])
    plt.show()
```

```python
for column in binominal:
    graf_for_binominal_churn(df, column)
```

Посмотрим на количественные признаки.

```python
# для количественных признаков сделаем функцию,
# которая будет строить гистограмму и боксплот с разбивкой по факту оттока

def graf_for_numerical_churn(data, column):   
    pallete = sns.color_palette("tab10", 7)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (14,6), gridspec_kw={'width_ratios': [2, 1]})
    min_value = min(df[column])
    max_value = max(df[column])
    bins_range = np.linspace(min_value, max_value, 15) # список границ столбиков гистограмм
    sns.histplot(data=df.query('churn == False')[column], color=pallete[0], 
                 stat='probability', common_norm=False, ax=ax1, bins=bins_range) 
    sns.histplot(data=df.query('churn == True')[column], color=pallete[1], 
                 stat='probability', common_norm=False, ax=ax1, bins=bins_range)
    ax1.set_ylabel('Частота признака')
    ax1.set_xlabel('{}'.format(column))
    ax1.set_title('Распределение признака {}'.format(column))
    ax1.legend(labels=['Остались', 'Ушли'])
    
    sns.boxplot(y=column, x='churn', data=df, ax=ax2, palette="deep")
    ax2.set_title('Распределение признака {}'.format(column))
    ax2.set_xlabel(' ')
    ax2.set_ylabel('{}'.format(column))
    ax2.set_xticklabels(['Остались', 'Ушли'])
    plt.show()
```

### Построение модели прогнозирования оттока клиентов

#### Подготовка данных к обучению модели

Построим модель бинарной классификации клиентов, где целевой признак — факт оттока клиента в следующем месяце (`churn`). Для начала разделим наши данные на признаки (матрица X) и целевую переменную (y). Разделим модель на обучающую и валидационную выборки путем случайного разбиения методом train_test_split() в пропорции 80% на 20%, после чего проведем стандартизацию.

```python
X = df_for_corr.drop(['churn'], axis=1)
y = df_for_corr['churn']
```

```python
# разделяем модель на обучающую и валидационную выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, stratify=y)
```

```python
# стандартизируем данные методом StandartScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train_st = scaler.transform(X_train)
X_test_st = scaler.transform(X_test)
```

#### Обучение модели логистической регрессией и случайным лесом

Построим модель логистической регрессии:

```python
# создадим алгоритм для модели логистической регрессии
model_lr = LogisticRegression(solver='liblinear', random_state=0)

# обучим модель
model_lr.fit(X_train_st, y_train)

# воспользуемся обученной моделью, чтобы сделать прогнозы
predictions_lr = model_lr.predict(X_test_st)
probabilities_lr = model_lr.predict_proba(X_test_st)[:,1]
```

Построим модель случайного леса:

```python
# создадим алгоритм для модели логистической регрессии
model_rf = RandomForestClassifier(n_estimators = 42, random_state = 0)

# обучим модель
model_rf.fit(X_train_st, y_train)

# воспользуемся обученной моделью, чтобы сделать прогнозы
predictions_rf = model_rf.predict(X_test_st)
probabilities_rf = model_rf.predict_proba(X_test_st)[:,1]
```

Зададим функцию, которая позволит нам вывести метрики.

```python
def print_all_metrics(y_true, y_pred, y_proba, title):
    print('\033[1m' + title + '\033[0m')
    print('\tAccuracy: {:.2f}'.format(accuracy_score(y_true, y_pred)))
    print('\tPrecision: {:.2f}'.format(precision_score(y_true, y_pred)))
    print('\tRecall: {:.2f}'.format(recall_score(y_true, y_pred)))
```

```python
print_all_metrics(y_test, predictions_lr, probabilities_lr, 'Метрики модели логистической регрессии:')
print_all_metrics(y_test, predictions_rf, probabilities_rf, 'Метрики модели случайного леса:')
```

Обе модели в целом леса показывают приличную долю правильных ответов и хорошо справляются с задачей прогнозирования оттока клиентов. Произведем оценку важности признаков для каждой модели, посмотрим есть ли какие-то признаки, которые определены как важные обеими моделями.

#### Определение важности признаков, влияющих на отток

Создадим функцию, которая покажет нам важность признаков в каждой из моделей.

```python
def feature(corrdf, model_feature, reg):
    feats = {}
    for feature, importance in zip(corrdf.columns, model_feature):
        feats[feature] = importance
    importances = pd.DataFrame.from_dict(feats, orient='index').rename(columns={0: 'Importance'})
    importances = importances.sort_values(by='Importance', ascending=False)
    importances = importances.reset_index()
    importances = importances.rename(columns={'index': 'Features'})
    sns.set(font_scale = 5)
    sns.set(style="whitegrid", color_codes=True, font_scale = 1.7)
    fig, ax = plt.subplots()
    fig.set_size_inches(30,15)
    sns.barplot(x=importances['Importance'], y=importances['Features'], data=importances, color='skyblue')
    plt.xlabel('Importance', fontsize=25, weight = 'bold')
    plt.ylabel('Features', fontsize=25, weight = 'bold')
    plt.title(f'Feature Importance {reg}', fontsize=25, weight = 'bold')
    plt.show()
    display(importances.head())
```

```python
feature(X, abs(model_lr.coef_[0]), 'LogisticRegression')
display(feature(X, model_rf.feature_importances_, 'RandomForestClassifier'))
```

### Гипотеза о различии доходов между клиентами с разным количеством продуктов

Для проверки гипотез необходимо посмотреть на распределение исследуемого признака, перцентили этого значения по всем данным, а также на наличие аномальных значений.

```python
# выделим выборки в отдельные переменные 
clients_product_1 = df.query('products == 1')
clients_product_2 = df.query('products == 2')
```

Проверим получившиеся выборки с данными на нормальность распределения при помощи гистограмм и QQ-графиков.

Поскольку в выборках по зарплате разных групп клиентов распределение значений далеко от нормального и в данных имеются выбросы, а кроме того, вторая выборка заметно меньше первой, для выявления статистической значимости различий в зарплате `salary` будем применять непараметрический тест Манна-Уитни.

Зададим уровень значимости: 𝛼 = 0.05

Если p-value будет меньше выбранного уровня значимости 𝛼 = 0.05, то нулевую гипотезу об отсутствии отличий в зарплате отвергнем. Иначе сделаем вывод, что данные не позволили нам отвергнуть нулевую гипотезу.

**Введем основную и альтернативные гипотезы для проверки разности выборок:**

**H0:** Доходы клиентов, которые пользуются двумя продуктами, и теми, кто пользуются одним продуктом, не различаются.      
**H1:** Доходы клиентов, которые пользуются двумя продуктами, и теми, кто пользуются одним продуктом, различаются.

```python
alpha = 0.05  # критический уровень статистической значимости

results = stats.mannwhitneyu(clients_product_1['salary'], clients_product_2['salary'], alternative = 'two-sided')

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print('Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя')
```

Таким образом, на имеющихся данных, на уровне значимости 5% отвергаем нулевую гипотезу, согласно которой доходы клиентов, которые пользуются одним и двумя продуктами банка, не различается.     

### Гипотеза о различии оттока клиентов с наличием и отсутствием кредитной карты

Сформируем таблицу `df_card`, содержащую сведения об оттоке клиентов при наличии кредитной карты и без кредитной карты, после чего посмотрим имеются ли статистически значимые различия между между клиентами.

**Введем основную и альтернативные гипотезы для проверки разности выборок:**

**H0:** Доля оттока среди клиентов, имеющих и не имеющих кредитную карту, не различаются.     
**H1:** Доля оттока среди клиентов, имеющих и не имеющих кредитную карту, различаются.

При проверке гипотез мы имеем дело с бинарными данными, поэтому применим ztest. Количество наблюдений достаточно.

Зададим уровень значимости: 𝛼 = 0.05

Если p-value будет меньше выбранного уровня значимости 𝛼 = 0.05, то нулевую гипотезу отвергнем. Иначе сделаем вывод, что данные не позволили нам отвергнуть нулевую гипотезу.

Создадим функцию, которая позволит нам провести тестирование каждого из событий в исследуемых группах посредством z-теста, формула которого следующая:

$ z = \frac{p_1 - p_2}{\sqrt[2]{p*(1-p)*\big(\frac{1}{n_1} + \frac{1}{n_2}\big)}} $
</div>

```python
def st_test(test0, test1, total0, total1):  
    
    alpha = 0.05  # критический уровень статистической значимости
    
    # считаем пропорцию успехов в первой группе:
    p1 = test0 / total0 
    
    # считаем пропорцию успехов во второй группе:
    p2 = test1 / total1 
    
    # считаем пропорция успехов в комбинированном датасете:
    p_combined = (test0 + test1) / (total0 + total1) 
    
    # считаем разницу пропорций в датасетах
    difference = p1 - p2 
    
    # считаем статистику в стандартных отклонениях стандартного нормального распределения
    z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/total0 + 1/total1))
    
    # задаем стандартное нормальное распределение (среднее 0, стандартное отклонение 1)
    distr = stats.norm(0, 1) 
    
    p_value = (1 - distr.cdf(abs(z_value))) * 2
    
    return p_value
```

```python
cred_card = df.pivot_table(index='credit_card', columns='churn', values='userid', aggfunc='count')
cred_card = cred_card.rename_axis(None, axis="columns").reset_index()
cred_card['total'] = cred_card.sum(axis=1)
cred_card
```

```python
alpha = 0.05  # критический уровень статистической значимости

results = st_test(cred_card.iloc[0, 2], cred_card.iloc[1, 2],
                  cred_card.iloc[0, 3], cred_card.iloc[1, 3])

print('p-значение: ', results)

if results < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print('Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя')
```

Таким образом, на имеющихся данных, на уровне значимости 5% отвергаем нулевую гипотезу, согласно которой доля оттока среди клиентов, имеющих и не имеющих кредитную карту, не различаются.

Стоит предложить этим клиентам кредитную карту на сопоставимых с конкурентами условиях, а также рассмотреть возможность предложения смешанного продукта: депозит + кредитная карта на более выгодных условиях, чем у клиента есть сейчас при условии поддержания остатка на депозите до определенного уровня. Кроме того, стоит рассмотреть возможность предложения клиентам кредит на крупные покупки (в т.ч. автомобиль). Способ информирования: смс, банковское приложение, обзвон.

**Группа 3 (баланс на счете больше 350000; возраст менее 60 лет; скоринговый балл более 640; имеется кредитная карта):** клиенты с 1 продуктом, имеют высокие остатки на счетах, имеется кредитная карта и высокие показатели скорингового балла, а также среднюю заработную плату. При этом отток наблюдается в той части группы, в которой имеется активность. Количество клиентов - 1396, уровень оттока: 24.28%

***Рекомендации по снижению оттока:*** уровень оттока на 70% превышает средние показатели. Клиенты имеют кредитную карту, при этом 100% из тех, кто был активным пользователем карты уходит из Банка. Стоит предложить этим клиентам улучшить условия по кредитной карте (например, увеличение льготного периода, уменьшение процентной ставки, предусмотреть возможность снятия наличных без комиссии и т.п.), а также предложить смешанные продукты: депозит + кредитная карта или предложить кредит на крупные покупки (в т.ч. ипотеку) на выгодных условиях или инвестиционные продукты, имеющиеся в Банке. Способ информирования: смс, банковское приложение, выборочный обзвон.

**Группа 4**: клиенты с 3 и 4 продуктами. Практически все из них являются активными пользователями и имеют одним из продуктов кредитную карту, имеют высокий баланс по счету и высокий уровень заработной платы, а также высокий скоринговый балл. Вероятно причиной оттока могли послужить более выгодные условия, предлагаемые конкурентами, к примеру более худшие условия по комплексным продуктам (депозит + кредитная карта или депозит + дебетовая карта и т.п.). Количество клиентов - 314, уровень оттока: 100%.

***Рекомендации по снижению оттока:*** к этим клиентам необходимо отнестись максимально пристально, поскольку они итак уже пользуются гораздо большим количеством продуктов Банка, чем средний клиент. 
Вероятнее всего многие услуги, которыми пользуется клиент предлагают и другие Банки, необходимо посмотреть есть ли среди тех, которые предлагает наш Банк комплексные (кредит + кредитная карта, депозит + кредитная карта, инвестиционный счет + кредитная карта и т.п.) и как давно клиент ими пользуется, если условия по этому продукту давно не менялись и сильно отличаются от текущих предложений конкурентов, стоит предложить клиенту улучшения условий обслуживания по этим продуктам. 
Способ информирования: смс, банковское приложение, желательно обзвон максимально возможного количества клиентов с целью получения в том числе обратной связи.

**Общие рекомендации по снижению оттока:**

Если у клиента не указан баланс, и он является потребителем услуги, не предполагающей наличие дебетового счета, то он менее склонен к оттоку. Если у них долгосрочный банковский продукт, такие клиенты остаются с банком на довольно продолжительный срок, поэтому есть отличная возможность заслужить лояльность таких клиентов, например предложив им продукты, предполагающие наличие дебетового счета (вклады, дебетовые карты и т.д.). 
В целом, подавляющее большинство клиентов - это владельцы 1-2 продуктов, поэтому следует направить внимание на увеличение числа продуктов для уже имеющихся клиентов (с 1 продуктом), так как они по большей части лояльны к банку (клиенты с двумя продуктами самые лояльные) и более охотно воспользуются дополнительными продуктами, чем те, кто еще не является клиентом банка. Наибольший доход банку приносят именно такие клиенты.   

Какими методами можно улучшить продажи продукта, предлагая лучшие условия лояльным клиентам:
- клиент оформил кредитный продукт (ипотека/кредит наличными) или кредитную карту, в таком случае ему необходимо дополнительно предложить выгодную дебетовую карту с повышенным кэшбэком и возможно бесплатным обслуживанием;
- клиент оформил себе дебетовый продукт (депозит, дебетовая карта, инвестиционный счет и т.п.), в таком случае ему необходимо предложить кредитную карту с выгодными условиями (например, повышенным льготным периодом, возможностью бесплатного снятия наличных в каком-то лимите и т.п.), выгодные условия рефинансирования кредита (если у него они имеются в другом банке).

Примеры реализации принципов взаимодействия с клиентами и их удержания в виде маркетинговых действий:
- Взаимодействие в приложении банка и социальных медиа: рекомендательный и полезный контент для клиентов, включающий разъяснения для клиентов по новым продуктам.
- Персонализация контента: следует предлагать продукты только тем, кому они потенциально могли быть интересны, т.е. умеренное СМС-информирование / push уведомления в приложении.
- Клиенты должны получать примерно одинаковый уровень сервиса и набора услуг вне зависимости от выбранного канала связи с банком (как дистанционного, так и очного).

- на имеющихся данных, на уровне значимости 5% доля оттока среди клиентов, имеющих и не имеющих кредитную карту различается.
