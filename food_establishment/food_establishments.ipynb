#### Особенности сетевых объектов (число посадочных мест)

Воспользуемся функцией pd.cut для выделения заведений по типам в зависимости от количества посадочных мест, задав параметр bin, затем применим указанную функцию для создания нового столбца из значений столбца `number_of_seats`.

```python
# создадим список категорий
cut_labels = ['Небольшой (без посадочных мест)', 'Небольшой (с посадочными местами)', 'Средний', 'Крупный', 'Огромный']

# создадим список диапазонов значений применительно к каждой категории, указав границы каждого из них
cut_bins = [-1, 0, 50, 100, 300, 2000]
```

```python
# создадим новый столбец с категориями и применим к нему необходимые вычисления
rest_data['number_of_seats_category'] = pd.cut(rest_data['number_of_seats'], bins=cut_bins, labels=cut_labels)
```

```python
# проверим корректность произведенных преобразований
display(rest_data.head())
```

Объекты, которые не попали ни в одну из категорий, можно отнести к небольшим, но без посадочных мест, поскольку они вероятнее всего ориентированы на доставку или занимаются продажей фаст-фуда.

```python
# проверим отсутствие незаполненных значений в столбце
rest_data['number_of_seats_category'].isna().sum()
```

Сделаем таблицу сетевых заведений, посчитаем количество заведений в сети и медианное значение посадочных мест.

```python
chain_objects = (rest_data[rest_data['chain']=='да'].groupby('object_name')
                .agg({'object_name':'count', 'number_of_seats':'median'}))

chain_objects.columns = ['amount_objects', 'median_number_of_seats']
chain_objects.head(10)
```

Поскольку в таблице имеются сети, в которых указан всего 1 объект общественного питания, вероятно либо этот объект ошибочно отнесен к сети либо у сети имеются заведения, которые находятся в других городах. Кроме того, имеются сети с количеством посадочных мест равным 0 (вероятнее всего это заведение работает как доставка либо это фаст-фуд, где еду/напитки предлагают только с собой).     
Попробуем исключить эти сети из таблицы, для чего выделим в отдельную таблицу сети, в которых больше 1 заведения и больше 0 посадочных мест.

```python
chain_objects_clear = (chain_objects.copy(deep=True)
                       .query('amount_objects > 1 & median_number_of_seats > 0 ')
                       .astype('int16'))
chain_objects_clear = chain_objects_clear.sort_values('amount_objects', ascending=False)
chain_objects_clear.sort_values(['amount_objects', 'median_number_of_seats'])
```

#### Топ-10 улиц по количеству объектов общественного питания

Для дальнейшего исследования выделим название улицы из столбца с адресом, для чего сначала применим регулярное выражение.

```python
words = ['улица','ул','переулок','шоссе','проспект', 'просека', 'площадь','проезд', 'деревня', 'поселок'
         'село','аллея','бульвар','набережная','тупик','линия', 'поселение', 'квартал', 'микрорайон','просек', 'километр']

str_pat = r".*,\s*\b([^,]*?(?:{})\b[^,]*)[,$]+".format("|".join(words))

rest_data['street'] = rest_data['address'].str.extract(str_pat, flags=re.I)

print('В столбце с данными по улице {} пропущенных значений'.format(rest_data.street.isna().sum()))
```

Для тех адресов, из которых не удалось выделить адрес, создадим функцию.

```python
def get_street(value):
    # функция разобьет словосочетание по запятым 
    string = value.split(",")
    
    # если первым словосочетанием будет город Москва,
    # а вторым словом будем город - выделим третью часть, остальные в зависимости от длины адреса
    if "город" in string[0]:
        return re.sub(r'^\s+|\s+$', '', string[1])
    
    # если не подходит - выделим первую часть
    else:
        return re.sub(r'^\s+|\s+$', '', string[0])
```

Применим функцию к столбцу с данными об адресе и посмотрим имеются ли пропущенные значения после применения функции.

```python
rest_data['street'] = rest_data['address'].apply(get_street)
print('В столбце с данными по улице {} пропущенных значений'.format(rest_data.street.isna().sum()))
```
